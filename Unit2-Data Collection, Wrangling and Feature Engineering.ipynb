#### Memory usage optimization:
- float64 ‚Üí float32
- int64 ‚Üí int32 / int16
- object ‚Üí category

```python
import pandas as pd
import numpy as np

def reduce_memory_usage(df_orig: pd.DataFrame, verbose=True) -> pd.DataFrame:
    """
    Automatically optimize DataFrame memory usage:
    - Downcast int/float columns
    - Convert object columns to category (if the ratio of unique values < 50%)
    - Display memory usage before and after optimization
    """
    df = df_orig.copy()
    start_mem = df.memory_usage(deep=True).sum() / 1024**2
    
    for col in df.columns:
        col_type = df[col].dtype

        if pd.api.types.is_numeric_dtype(col_type):
            # Process integers
            if pd.api.types.is_integer_dtype(col_type):
                df[col] = pd.to_numeric(df[col], downcast="integer")
            # Process floats
            elif pd.api.types.is_float_dtype(col_type):
                df[col] = pd.to_numeric(df[col], downcast="float")
        
        # Process objects
        elif col_type == "object":
            num_unique = df[col].nunique()
            num_total = len(df[col])
            if num_unique / num_total < 0.5:  
                df[col] = df[col].astype("category")
    
    end_mem = df.memory_usage(deep=True).sum() / 1024**2
    
    if verbose:
        print(f"Memory usage before optimization: {start_mem:.2f} MB")
        print(f"Memory usage after optimization: {end_mem:.2f} MB")
        print(f"has reduced {(start_mem - end_mem) / start_mem * 100:.1f}%")
    
    return 
reduce_memory_usage(cashflow_annual_df)
```

### üé¨ Lesson 2.3: Perform Data Merging

```python
companies_info_df.head(3)
```

```python
industries_info_df.head(3)
```

```python
companies_industries_df = pd.merge(companies_info_df, industries_info_df, on='IndustryId', how='left')
companies_industries_df.head(3)
```

```python
eps_df = pd.read_csv(r'./data/unit2/eps_df.csv')
eps_df.head(3)
```

```python
cashflow_df = cashflow_df.drop(columns = ['Report Date', 
                                          'Currency', 
                                          'Publish Date', 
                                          'Restated Date', 
                                          'Shares (Basic)', 
                                          'Shares (Diluted)', 
                                          'Depreciation & Amortization'])
balance_df = balance_df.drop(columns = ['Report Date', 
                                        'Currency', 
                                        'Publish Date', 
                                        'Restated Date', 
                                        'Shares (Basic)', 
                                        'Shares (Diluted)'])
```

```python
# Find join key columns between dataframes
df_pair = [(companies_industries_df, income_df), (income_df, cashflow_df), \
           (cashflow_df, balance_df), (balance_df, eps_df)]
for df1, df2 in df_pair:
    print("dataframe has common columns: {}".format((set(df1.columns) & set(df2.columns))))
```

```python
merged_df = companies_industries_df.merge(income_df, on = ['Ticker','SimFinId'], how = 'left')\
                                   .merge(cashflow_df, on = ['Ticker','SimFinId', 'Fiscal Year','Fiscal Period'], how = 'left')\
                                   .merge(balance_df, on = ['Ticker','SimFinId','Fiscal Year','Fiscal Period'], how = 'left')\
                                   .merge(eps_df, on = ['Ticker','Fiscal Year','Fiscal Period'], how = 'left')
merged_df.head(5)
```

```python
merged_df = merged_df.drop(columns = ['SimFinId','IndustryId'])
merged_df.shape
```

```python
# Verify if the ticker unique value counts in merged_df is the same as in companies_indusries_df while using left join
assert len(merged_df['Ticker'].unique().tolist()) == \
len(companies_industries_df['Ticker'].unique().tolist())
```
